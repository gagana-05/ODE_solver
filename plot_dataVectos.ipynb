```python
x = []
y = []
z = []
#tried plotting roughly around 6000 pts for each of the 3 lists
#and normalized the data by dividing each element by 4096 
with open('5000pts.txt', 'r') as f:
    for line in f:
        first, second, x_comp, y_comp, z_comp = line.split()
        x.append(int(x_comp)/4096)
        y.append(int(y_comp)/4096)
        z.append(int(z_comp)/4096)

# print(x)
# print(y)
# print(z)
```

```python
import numpy as np

x = np.array(x)
y = np.array(y)
z = np.array(z)
print(x)
```

```python
# x = np.array([x])
# y = np.array([y])
# z = np.array([z])
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z, 'steelblue')
```
